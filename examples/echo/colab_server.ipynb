# ColabBridge — Echo Server

Minimal demo: any message sent over WebSocket is echoed back.

**Steps:**
1. Replace `YOUR_REGISTRY_URL` below with your deployed registry URL
2. Run Cell 1 (install), then Cell 2 (start server)
3. Open `examples/echo/web/index.html` in your browser
4. Paste your registry URL and click Connect

```python
# Cell 1 — Install ColabBridge
!pip install -q colabbridge-server
```

```python
# Cell 2 — Start echo server
from colabbridge import ColabBridge

bridge = ColabBridge(
    registry_url="YOUR_REGISTRY_URL",  # e.g. https://nova-url-registry-production.up.railway.app
)

@bridge.websocket("/ws")
def echo(data: bytes) -> dict:
    text = data.decode("utf-8", errors="replace")
    return {"echo": text, "length": len(data)}

@bridge.post("/ping")
def ping(message: str = "hello") -> dict:
    return {"pong": message}

# Prints the public URL and registers it with the registry.
# The web demo will auto-connect via the registry.
bridge.run()
```
